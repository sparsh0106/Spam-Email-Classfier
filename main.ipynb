# Importing the libraries

```python
import pandas as pd
import string

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

# Importing the dataset

```python
df = pd.read_csv('spam.csv')
df
```

# Checking for null values and dropping those rows

```python
df.isnull().sum()
df.dropna()
```

# Data preprocessing

```python
df['Message'] = df['Message'].str.replace(r"Subject:.*?(\r\n|\n)", "", regex=True)
df['Message'] = df['Message'].str.replace(f"[{string.punctuation}]", "", regex=True)
df['Message'] = df['Message'].str.replace(r'[\n\r]', " ", regex=True)
df['Category'] = df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
df['Message'] = df['Message'].str.replace(r'[^\x00-\x7F]+', '', regex=True)
df['Message'] = df['Message'].str.replace(r'[^a-zA-Z]', ' ', regex=True)
df['Message'] = df['Message'].str.replace(r'\s+', ' ', regex=True).str.strip()
df['Message'] = df['Message'].str.replace(r'http\S+|www\S+', '', regex=True)
df['Message'] = df['Message'].str.replace(r'http\S+|www\S+', '', regex=True)
df['Message'] = df['Message'].str.lower()

df
```

# Defining X and y columns

```python
X = df['Message']
y = df['Category']
```

# Splitting training and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

# Vectorizing the data and fitting training and testing data

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

# Initializing the mode and ftting training data in the model

```python
model = SVC(kernel='linear')
model.fit(X_train_features, y_train)
```

# Making prediction on test data

```python
predictions = model.predict(X_test_features)
```

# Getting the accuracy score of the model

```python
accuracy = accuracy_score(y_test, predictions)
print(f"Model Accuracy: {accuracy*100:.2f}%")
```
